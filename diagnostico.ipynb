```python
#Problema 1

def more_retweeted(list_tweets):
    sorted_tweets = sorted(list_tweets, key=lambda tweet : tweet['retweetCount'])
    top = list()
    for i in range(10):
        mayor = sorted_tweets[len(sorted_tweets)-1-i]
        top.append(mayor)
    for i in range(len(top)):
        print(i+1, top[i]['content'])
        
#more_retweeted(tweets)
```

```python
# Problema 2

def users_more_retweeted(list_tweets):
    users = list()
    for t in range(len(list_tweets)):
        user = list_tweets[t]['user']
        users.append(user)
        
    total_users = list()
    total_username = list()
    for i in range(len(users)):
        username = users[i]['username']
        if username not in total_username:
            total_username.append(username)
            total_users.append(users[i])
    
    sorted_tweets = sorted(total_users, key=lambda user : user['statusesCount'])
        
    top = list()
    for i in range(10):
        mayor = sorted_tweets[len(sorted_tweets)-1-i]
        top.append(mayor)
    for i in range(len(top)):
        print(top[i]['username'])
        
#users_more_retweeted(tweets)
```

```python
# Problema 3

import operator

def top_days(list_tweets):
    days = dict()
    for t in range(len(list_tweets)):
        date = list_tweets[t]['date'] 
        key_date = date[0:10]
        if key_date not in list(days.keys()):
            days[key_date] = 1
        else:
            days[key_date] += 1
            
    days_sort = sorted(days.items(), key=operator.itemgetter(1), reverse=True)
    
    for i in range(10):
        day = days_sort[i][0]
        tweets = days_sort[i][1]
        print(i+1, day, "n√∫mero de tweets:", tweets)
        
#top_days(tweets)
```
